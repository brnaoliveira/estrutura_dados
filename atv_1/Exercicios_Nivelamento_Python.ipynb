##**Questão 1:**
Analisando matematicamente as mortes do 7 reinos, o conselheiro Robert descobriu
que elas acontecem exponencialmente da seguinte forma: As mortes acontecem
exponencialmente de acordo com o número de famílias envolvidas nas batalhas e o
tempo. De acordo com o diagrama abaixo. Portanto, escreva um programa no qual a
entrada **n** representa o número de família e o resultado seja a seguinte tabela:

1 (1 família, nenhuma morte)

2 4 (2 familias, 4 mortes no primeiro ano)

3 6 9 (3 familias, 6 mortes no primeiro ano, 9 mortes no segundo ano)

4 8 12 16

n...

obs: o programa deve imprimir somente a tabela, sem os comentários

```python
def mortes_reino(n):
    tabela = []
    for i in range(1, n + 1):
        linha = []
        mortes = i
        for j in range(i):
            linha.append(mortes)
            mortes += i
        tabela.append(linha)

    for linha in tabela:
        for elemento in linha:
            print(elemento, end=" ")
        print()
```

```python
n = int(input("Digite o número de famílias: "))
mortes_reino(n)
```

##**Questão 2:**
Dada uma String, encontre o tamanho da maior substring sem repetir
caracteres

Exemplo1 :

**Input**: "abcabcbb"

**Output**: 3

Explicação: A resposta é "abc", de tamanho 3.

Example 2:

**Input**: "bbbbb”

**Output**: 1

Explicação: A resposta é "b", de tamanho 1.

Example 3:

**Input**: "pwwkew"

**Output**: 3

Explicação: A resposta é "wke", de tamanho 3.


```python
def tamanho_maior_substring(s):
    tokens = []
    tokens_sem_repeticoes = []

    for i in range(len(s)):
      tokens.append(s[i])

    for letra in tokens:
      if letra not in tokens_sem_repeticoes:
        tokens_sem_repeticoes.append(letra)


    maior = len(tokens_sem_repeticoes)



    return maior
```

```python
s = input("Digite a string: ")
print(tamanho_maior_substring(s))
```

##**Questão 3:**
Desenvolva um jogo da forca. O programa terá uma lista de palavras lidas de uma lista e escolherá uma palavra aleatoriamente. O jogador vai tentar descobrir essa palavra, mas só poderá errar 6 vezes antes de ser enforcado.

Digite uma letra: A
-> Você errou pela 1ª vez. Tente de novo!

Digite uma letra: O
A palavra é: _ _ _ _ O

Digite uma letra: E
A palavra é: _ E _ _ O

Digite uma letra: S
-> Você errou pela 2ª vez. Tente de novo!

Quando errar pela 6ª vez imprimir: Enforcado e exibir a palavra correta

```python
import random

def jogo_da_forca():
    palavras = ["ABACAXI", "BANANA", "MORANGO", "LARANJA", "UVA"]
    palavra_secreta = random.choice(palavras)
    letras_descobertas = ["_"] * len(palavra_secreta)
    tentativas = 0

    print("Jogo da Forca")
    print("Adivinhe a palavra secreta!")
    print("Dica: é uma fruta")
    print()

    while tentativas < 6:
        letra = input("Digite uma letra: ").upper()

        if letra in palavra_secreta:
            for i in range(len(palavra_secreta)):
                if palavra_secreta[i] == letra:
                    letras_descobertas[i] = letra
        else:
            tentativas += 1
            print(f"Você errou pela {tentativas}ª vez. Tente de novo!")

        print("A palavra é:", " ".join(letras_descobertas))
        print()

        if "_" not in letras_descobertas:
            print("Parabéns! Você acertou a palavra!")
            return

    print("Enforcado!")
    print("A palavra correta era:", palavra_secreta)

jogo_da_forca()
```

##**Questão 4:**

Desenvolva um sistema de banco em python. Você deve exibir 6 menus:

1. Criar nova conta. Uma nova conta deve conter ID, NOME e SALDO
2. Adicionar dinheiro a conta. Adiciona uma quantia (dita pelo usuário) a conta de um ID especifico, mostra quando ficou o saldo final.
3. Remover dinheiro da conta. Remove uma quantia (dita pelo usuário) a conta de um ID especifico, mostra quanto ficou o saldo final. Emitir um alerta caso o saldo tenha ficado negativo.
4. Mostrar saldo de um nome dado
5. Apagar uma conta (apaga todos os dados de um determinado ID)
6. Apagar todas as contas

ps: lembre-se de tratar exceções, caso o menu selecionado,  ID ou nome sugerido não exista imprimir uma alerta avisando que não existe.



```python
class ContaBancaria:
    def __init__(self, id_conta, nome, saldo):
        self.id_conta = id_conta
        self.nome = nome
        self.saldo = saldo

    def adicionar_dinheiro(self, quantia):
        self.saldo += quantia

    def remover_dinheiro(self, quantia):
        if self.saldo - quantia < 0:
            print("Saldo insuficiente!")
        else:
            self.saldo -= quantia

    def mostrar_saldo(self):
        return self.saldo


class SistemaBancario:
    def __init__(self):
        self.contas = {}

    def criar_conta(self, id_conta, nome, saldo):
        if id_conta in self.contas:
            print("Já existe uma conta com esse ID.")
        else:
            conta = ContaBancaria(id_conta, nome, saldo)
            self.contas[id_conta] = conta
            print("Conta criada com sucesso.")

    def adicionar_dinheiro_conta(self, id_conta, quantia):
        if id_conta in self.contas:
            conta = self.contas[id_conta]
            conta.adicionar_dinheiro(quantia)
            print(f"Adicionado {quantia} ao saldo da conta {id_conta}. Saldo atual: {conta.mostrar_saldo()}")
        else:
            print("A conta não existe.")

    def remover_dinheiro_conta(self, id_conta, quantia):
        if id_conta in self.contas:
            conta = self.contas[id_conta]
            conta.remover_dinheiro(quantia)
            print(f"Removido {quantia} do saldo da conta {id_conta}. Saldo atual: {conta.mostrar_saldo()}")
        else:
            print("A conta não existe.")

    def mostrar_saldo_nome(self, nome):
        for conta in self.contas.values():
            if conta.nome == nome:
                saldo = conta.mostrar_saldo()
                print(f"O saldo da conta de {nome} é {saldo}")
                return
        print("O nome não corresponde a nenhuma conta.")

    def apagar_conta(self, id_conta):
        if id_conta in self.contas:
            del self.contas[id_conta]
            print("Conta apagada com sucesso.")
        else:
            print("A conta não existe.")

    def apagar_todas_contas(self):
        self.contas.clear()
        print("Todas as contas foram apagadas.")

sistema = SistemaBancario()
```

```python
while True:
    print("\n------ MENU ------")
    print("1. Criar nova conta")
    print("2. Adicionar dinheiro a conta")
    print("3. Remover dinheiro da conta")
    print("4. Mostrar saldo de um nome dado")
    print("5. Apagar uma conta")
    print("6. Apagar todas as contas")
    print("0. Sair")
    opcao = input("Escolha uma opção: ")

    if opcao == "1":
        id_conta = input("Digite o ID da nova conta: ")
        nome = input("Digite o nome do titular da conta: ")
        saldo = float(input("Digite o saldo inicial da conta: "))
        sistema.criar_conta(id_conta, nome, saldo)

    elif opcao == "2":
        id_conta = input("Digite o ID da conta: ")
        quantia = float(input("Digite a quantia a ser adicionada: "))
        sistema.adicionar_dinheiro_conta(id_conta, quantia)

    elif opcao == "3":
        id_conta = input("Digite o ID da conta: ")
        quantia = float(input("Digite a quantia a ser removida: "))
        sistema.remover_dinheiro_conta(id_conta, quantia)

    elif opcao == "4":
        nome = input("Digite o nome do titular da conta: ")
        sistema.mostrar_saldo_nome(nome)

    elif opcao == "5":
        id_conta = input("Digite o ID da conta a ser apagada: ")
        sistema.apagar_conta(id_conta)

    elif opcao == "6":
        confirmacao = input("Tem certeza que deseja apagar todas as contas? (S/N): ")
        if confirmacao.upper() == "S":
            sistema.apagar_todas_contas()
        elif confirmacao.upper() == "N":
          print("Saindo do programa...")
          break


    elif opcao == "0":
        print("Saindo do programa...")
        break

    else:
        print("Opção inválida. Por favor, escolha uma opção válida.")

```

##**Questão5:**
Dado um array agrupe os anagramas

Example:

Input:
["eat", "tea", "tan", "ate", "nat", "bat"],

Output:

[ ["ate","eat","tea"],

["nat","tan"],

["bat"] ]


```python
def agrupar_anagramas(palavras):
    anagramas = {}

    for palavra in palavras:
        chave = "".join(sorted(palavra))
        if chave in anagramas:
            anagramas[chave].append(palavra)
        else:
            anagramas[chave] = [palavra]

    return list(anagramas.values())

palavras = ["eat", "tea", "tan", "ate", "nat", "bat"]
resultado = agrupar_anagramas(palavras)
print(resultado)
```
