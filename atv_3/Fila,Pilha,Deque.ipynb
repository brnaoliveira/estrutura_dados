#Questão 1#
Em sala de aula, discutimos que os TADs Pilha, Fila, e Deque podem ser considerados casos específicos do TAD Sequência. Em outras palavras, ao restringirmos a forma que uma Sequência é manipulada, teremos o comportamento esperado de uma Pilha, Fila, ou Deque. Nesta questão, implemente os TADs Pilha, Fila, e Deque com base em listas.

a) Pilha – Operações Push, Pop, top, isEmpty, size

b) Fila – Operações Queue,Dequeue, peek, isEmpty, size

c) Deque – PushBack, PushFront, PopBack, PopFront,Back, Front

```python
class Pilha:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()

    def top(self):
        if not self.is_empty():
            return self.items[-1]

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)
```

```python
# Testando a Pilha
pilha = Pilha()
print(pilha.is_empty())
pilha.push(1)
pilha.push(2)
pilha.push(3)
print(pilha.size())
print(pilha.top())
print(pilha.pop())
```

```python
class Fila:
    def __init__(self):
        self.items = []

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        if not self.is_empty():
            return self.items.pop(0)

    def peek(self):
        if not self.is_empty():
            return self.items[0]

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)
```

```python
# Testando a Fila
fila = Fila()
print(fila.is_empty())
fila.enqueue(1)
fila.enqueue(2)
fila.enqueue(3)
print(fila.size())
print(fila.peek())
print(fila.dequeue())
print(fila.size())
```

```python
class Deque:
    def __init__(self):
        self.items = []

    def push_back(self, item):
        self.items.append(item)

    def push_front(self, item):
        self.items.insert(0, item)

    def pop_back(self):
        if not self.is_empty():
            return self.items.pop()

    def pop_front(self):
        if not self.is_empty():
            return self.items.pop(0)

    def back(self):
        if not self.is_empty():
            return self.items[-1]

    def front(self):
        if not self.is_empty():
            return self.items[0]

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)
```

```python
# Testando o Deque
deque = Deque()
print(deque.is_empty())
deque.push_back(1)
deque.push_back(2)
deque.push_front(3)
print(deque.size())
print(deque.back())
print(deque.front())
print(deque.pop_back())
print(deque.pop_front())
print(deque.size())
```

#Questão 2:
Suponha que você já possui uma estrutura Pilha implementada corretamente. Reuse esta estrutura
Pilha para implementar uma estrutura Fila. Em outras palavras, implemente uma estrutura Fila usando
internamente uma ou mais estruturas Pilha. Lembre-se que as operações da Fila devem obedecer a estratégia
FIFO.

```python
class Fila:
    def __init__(self):
        self.entrada = Pilha()
        self.saida = Pilha()

    def enqueue(self, item):
        self.entrada.push(item)

    def dequeue(self):
        if self.saida.is_empty():
            # Transfere os elementos da pilha de entrada para a pilha de saída
            while not self.entrada.is_empty():
                self.saida.push(self.entrada.pop())
        return self.saida.pop()

    def peek(self):
        if self.saida.is_empty():
            # Transfere os elementos da pilha de entrada para a pilha de saída
            while not self.entrada.is_empty():
                self.saida.push(self.entrada.pop())
        return self.saida.top()

    def is_empty(self):
        return self.entrada.is_empty() and self.saida.is_empty()

    def size(self):
        return self.entrada.size() + self.saida.size()




fila = Fila()
fila.enqueue(1)
fila.enqueue(2)
fila.enqueue(3)
print(fila.dequeue())
print(fila.peek())
print(fila.is_empty())
print(fila.size())
```

#Questão 3:
Palíndromo é a frase ou palavra que mantém o mesmo sentido quando lida de trás pra frente. São
exemplos de palíndromo: "arara", "osso", "reler", "somos", e “amor à roma”. Podemos também considerar
palíndromos algumas combinações de palavras em que desprezamos pontuações, acentos e espaços em branco,
como por exemplo: “ralo do dólar”, “até o poeta”, “após a sopa”, etc.
Neste contexto:

a) Use uma Pilha para projetar uma função que recebe uma string como entrada e verifica se ela é ou não um palíndromo.

b) Use um Deque para projetar uma função que recebe uma string como entrada e verifica se ela é ou não um palíndromo.

```python
def verifica_palindromo_pilha(string):
    string = string.lower().replace(" ", "").replace(",", "").replace(".", "").replace("!", "").replace("?", "").replace("'", "")
    pilha = Pilha()
    for char in string:
        pilha.push(char)
    for char in string:
        if char != pilha.pop():
            return False
    return True


frase = "Ralamos o sabor da comida"
if verifica_palindromo_pilha(frase):
    print(f"A frase '{frase}' é um palíndromo.")
else:
    print(f"A frase '{frase}' não é um palíndromo.")

print()
frase2 = "Anotaram a data da maratona"
if verifica_palindromo_pilha(frase2):
    print(f"A frase '{frase2}' é um palíndromo.")
else:
    print(f"A frase '{frase2}' não é um palíndromo.")

```

```python
def verifica_palindromo_deque(string):
    string = string.lower().replace(" ", "").replace(",", "").replace(".", "").replace("!", "").replace("?", "").replace("'", "")
    deque = Deque()
    for char in string:
        deque.push_back(char)
    while deque.size() > 1:
        if deque.pop_front() != deque.pop_back():
            return False
    return True

frase = "Até o poeta"
if verifica_palindromo_deque(frase):
    print(f"A frase '{frase}' é um palíndromo.")
else:
    print(f"A frase '{frase}' não é um palíndromo.")



print()
frase2 = "Anotaram a data da maratona"
if verifica_palindromo_deque(frase2):
    print(f"A frase '{frase2}' é um palíndromo.")
else:
    print(f"A frase '{frase2}' não é um palíndromo.")

```

#Questão 4
Dado duas pilhas P1 e P2 que armazenam números inteiros, faça um programa que tenha as seguintes funções:
1. Uma função para testar se as duas pilhas P1 e P2 são iguais. Caso não forem iguais diga qual pilha é maior
2. Uma função que forneça o maior, o menor e a média aritmética dos elementos de uma pilha.
3. Uma função para transferir elementos de P1 para P2 (copia).
4. Uma função para retornar o número de elementos de uma pilha que possuem valor ímpar.
5. Uma função para retornar o número de elementos da uma pilha que possuem valor par

```python
def pilhas_iguais(P1, P2):
    if P1.size() != P2.size():
        return False
    while not P1.is_empty():
        if P1.pop() != P2.pop():
            return False
    return True


def maior_menor_media(pilha):
    maior = float('-inf')
    menor = float('inf')
    soma = 0
    qtd_elementos = pilha.size()

    while not pilha.is_empty():
        elemento = pilha.pop()
        soma += elemento
        if elemento > maior:
            maior = elemento
        if elemento < menor:
            menor = elemento

    media = soma / qtd_elementos
    return maior, menor, media


def copiar_pilha(P1, P2):
    while not P1.is_empty():
        elemento = P1.pop()
        P2.push(elemento)


def contar_elementos_impares(pilha):
    qtd_impares = 0
    aux = Pilha()

    while not pilha.is_empty():
        elemento = pilha.pop()
        if elemento % 2 != 0:
            qtd_impares += 1
        aux.push(elemento)

    copiar_pilha(aux, pilha)

    return qtd_impares


def contar_elementos_pares(pilha):
    qtd_pares = 0
    aux = Pilha()

    while not pilha.is_empty():
        elemento = pilha.pop()
        if elemento % 2 == 0:
            qtd_pares += 1
        aux.push(elemento)

    copiar_pilha(aux, pilha)

    return qtd_pares


pilha1 = Pilha()
pilha1.push(1)
pilha1.push(2)
pilha1.push(3)

pilha2 = Pilha()
pilha2.push(1)
pilha2.push(2)
pilha2.push(4)

if pilhas_iguais(pilha1, pilha2):
    print("As pilhas são iguais.")
else:
    if pilha1.size() > pilha2.size():
        print("A pilha 1 é maior.")
    else:
        print("A pilha 2 é maior.")

maior, menor, media = maior_menor_media(pilha1)
print(f"Maior elemento: {maior}")
print(f"Menor elemento: {menor}")
print(f"Média aritmética: {media}")
print()

pilha3 = Pilha()
pilha3.push(1)
pilha3.push(3)
pilha3.push(5)

pilha4 = Pilha()
pilha3.push(2)
pilha3.push(3)
pilha3.push(6)


copiar_pilha(pilha3, pilha4)

print(f"Número de elementos ímpares: {contar_elementos_impares(pilha4)}")
print(f"Número de elementos pares: {contar_elementos_pares(pilha4)}")
```

#Questão 5.
Crie um MÉTODO que retira todos os elementos ímpares da fila, mantendo a ordem original dos
elementos pares da fila. Os elementos pares devem ser removidos temporariamente para uma pilha e depois
retornados para a fila assim que os elementos ímpares forem removidos.

```python
def remove_elementos_impares(fila):
    pilha_auxiliar = []
    nova_fila = Fila()

    while not fila.is_empty():
        elemento = fila.dequeue()
        if elemento % 2 == 0:
            pilha_auxiliar.append(elemento)

    while not nova_fila.is_empty():
        nova_fila.dequeue()

    while pilha_auxiliar:
        nova_fila.enqueue(pilha_auxiliar.pop())

    return nova_fila


fila = Fila()
fila.enqueue(1)
fila.enqueue(2)
fila.enqueue(3)
fila.enqueue(4)
fila.enqueue(5)

fila_atualizada = remove_elementos_impares(fila)
while not fila_atualizada.is_empty():
    print(fila_atualizada.dequeue())
```
